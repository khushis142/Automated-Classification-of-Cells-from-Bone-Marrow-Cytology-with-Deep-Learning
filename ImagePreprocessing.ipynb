```python
import os
import random
from shutil import copy, make_archive
```

```python
data_root = '/kaggle/input/bone-marrow-cell-classification/bone_marrow_cell_dataset'

os.makedirs('/kaggle/working/dataset', exist_ok=True) # create a dataset folder to hold all the files that I wanted to download
```

```python
k = 1000 # randomly select 1000 images from data folders
```

```python
for d in ['EOS/0001-1000','NGB/0001-1000','NGS/0001-1000','LYT/0001-1000','MON/0001-1000']:
    # list all images in train/test folder
    dir_path = os.path.join(data_root, d)
    files = os.listdir(dir_path)
    
    # copy images to target folder
    target_dir = os.path.join('dataset', d)
    os.makedirs(target_dir, exist_ok=True) 
    for f in random.sample(files, k=k): # randomly select k images and copy them to the target folder
        src_file = os.path.join(dir_path, f)
        copy(src_file, target_dir)
```

```python
import sys
import cv2 
import numpy as np
#import os
import glob
```

```python
def mainFunction(input_directory,output_directory):
    directory = input_directory
    os.chdir(directory)

    for imgname in glob.glob("*.jpg"):
        #print(imgname)
        # Read your image here
        src = cv2.imread(imgname)

        # Check if image is loaded fine
        if src is None:
            print ('Error opening image!')
            print ('Usage: hough_circle.py [image_name -- default ' + imgname + '] \n')
            return -1
        
        # Process your image here
        #print(src.shape)
        src = cv2.resize(src, (224, 224))
        #print(src.shape)
        
        src = cv2.fastNlMeansDenoisingColored(src,None,10,10,7,21)
        
        sharpen_filter=np.array([[-1,-1,-1],
                 [-1,9,-1],
                [-1,-1,-1]])
        src=cv2.filter2D(src,-1,sharpen_filter)

        # Save your image here
        directory = output_directory
        os.chdir(directory) 
        filename = str(imgname)
        cv2.imwrite(filename, src)

        # Change the directory to the data folder, before read the next image
        directory = input_directory
        os.chdir(directory)
  
```

```python
os.makedirs('/kaggle/working/processed_EOS', exist_ok=True)
os.makedirs('/kaggle/working/processed_BAS', exist_ok=True)
os.makedirs('/kaggle/working/processed_HAC', exist_ok=True)
os.makedirs('/kaggle/working/processed_NGS', exist_ok=True)
os.makedirs('/kaggle/working/processed_NGB', exist_ok=True)
os.makedirs('/kaggle/working/processed_MON', exist_ok=True)
os.makedirs('/kaggle/working/processed_LYT', exist_ok=True)
```

```python
if __name__ == "__main__":
    output_directory= r'/kaggle/working/processed_EOS'
    input_directory= r'/kaggle/working/dataset/EOS/0001-1000'
    mainFunction(input_directory,output_directory)
    output_directory= r'/kaggle/working/processed_BAS'
    input_directory= r'/kaggle/working/dataset/BAS'
    mainFunction(input_directory,output_directory)
    output_directory= r'/kaggle/working/processed_HAC'
    input_directory= r'/kaggle/working/dataset/HAC'
    mainFunction(input_directory,output_directory)
    output_directory= r'/kaggle/working/processed_NGS'
    input_directory= r'/kaggle/working/dataset/NGS/0001-1000'
    mainFunction(input_directory,output_directory)
    output_directory= r'/kaggle/working/processed_NGB'
    input_directory= r'/kaggle/working/dataset/NGB/0001-1000'
    mainFunction(input_directory,output_directory)
    output_directory= r'/kaggle/working/processed_MON'
    input_directory= r'/kaggle/working/dataset/MON/0001-1000'
    mainFunction(input_directory,output_directory)
    output_directory= r'/kaggle/working/processed_LYT'
    input_directory= r'/kaggle/working/dataset/LYT/0001-1000'
    mainFunction(input_directory,output_directory)
```
